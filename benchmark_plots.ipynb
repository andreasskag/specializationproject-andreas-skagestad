```python
####plots speedup on different devices

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                '%1.2f' % (height),
                ha='center', va='bottom')

fig, ax = plt.subplots(figsize=figusize)
plt_n = 1
bredde = 0.25
x = np.arange(num_runs)

GTX1080TI_Speed = np.divide(I7_5930K,np.where(GTX1080TI > 0, GTX1080TI, 1e6))

GTX960_Speed = np.divide(I7_5930K, GTX960)
GTX750M_Speed = np.divide(I7_5930K, GTX750M)
for i in range(0,1):
    rect1 = ax.bar(x + bredde / 2, (GTX1080TI_Speed[num_runs*i:num_runs*(i + 1), plt_n]), label='GTX1080TI', width=bredde, log=False)
    rect2 = ax.bar(x - bredde / 2, (GTX960_Speed[num_runs*i:num_runs*(i + 1), plt_n]), label='GTX960', width=bredde, log=False)
    rect3 = ax.bar(x - 3*bredde / 2, (GTX750M_Speed[num_runs*i:num_runs*(i + 1), plt_n]), label='GTX750M', width=bredde, log=False)
    ax.legend()
plt.xticks(x, PARTICLES)
ax.set_xlabel('Number of particles')
ax.set_ylabel('Factor of speedup $S$')
autolabel(rect1)
autolabel(rect2)
autolabel(rect3)
ax.set_title('Speedup on different devices, 667 timesteps')    
plt.savefig(path + 'speed_long.png', dpi=DPI)
fig2, ax = plt.subplots(figsize=figusize)

for i in range(1,2):
    rect1 = ax.bar(x + bredde / 2, (GTX1080TI_Speed[num_runs*i:num_runs*(i + 1), plt_n]), label='GTX1080TI', width=bredde, log=False)
    rect2 = ax.bar(x - bredde / 2, (GTX960_Speed[num_runs*i:num_runs*(i + 1), plt_n]), label='GTX960', width=bredde, log=False)
    rect3 = ax.bar(x - 3*bredde / 2, (GTX750M_Speed[num_runs*i:num_runs*(i + 1), plt_n]), label='GTX750M', width=bredde, log=False)
    ax.legend()
plt.xticks(x, PARTICLES)
#ys = np.arange(2.*num_runs)
#ys[0:5] = GTX960_Speed[5:10,0];
#ys[5:10] = GTX750TI_Speed[5:10,0];
#plt.yticks(ys)
ax.set_xlabel('Number of particles')
ax.set_ylabel('Factor of speedup $S$')
ax.set_title('Speedup on different devices, 43 timesteps')
autolabel(rect1)
autolabel(rect2)
autolabel(rect3)
#plt.xscale('log')
plt.savefig(path + 'speed_short.png', dpi=DPI)
plt.show()
```
